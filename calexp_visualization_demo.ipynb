```python
from lsst.daf.butler import Butler
import lsst.afw.display as afwDisplay
```

```python
import matplotlib.pyplot as plt
```

```python
butler =  Butler('/repo/embargo', collections='LSSTComCamSim/runs/intermittentcumulativeDRP/20240402_03_04/d_2024_03_29/DM-43865')
```

```python
calexps = list(butler.registry.queryDatasets('calexp'))
```

```python
calexp = butler.get(calexps[0])
```

```python
sourceTable = butler.get("sourceTable", dataId=calexps[0].dataId)
```

```python
afwDisplay.setDefaultBackend('matplotlib')
plt.style.use('tableau-colorblind10')
# The following line makes the plots interactive. Mouse over the plot and you will see options pop up to zoom, pan, etc.
%matplotlib widget 
```

```python
fig = plt.figure(figsize=(6,6))
display = afwDisplay.Display(frame=fig)
display.scale('asinh', 'zscale')
display.mtv(calexp.image)
```

```python
# Same thing with sources plotted over the image
fig = plt.figure(figsize=(6,6))
display = afwDisplay.Display(frame=fig)
display.scale('asinh', 'zscale')
display.mtv(calexp.image)

plt.scatter(sourceTable['x'], sourceTable['y'], marker='+', color='r')
```
