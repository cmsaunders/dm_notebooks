# Accessing data processed through the Data Reduction Pipeline

```python
from lsst.daf.butler import Butler
import pandas as pd
pd.set_option('display.max_columns', None)
```

### The "butler" tool is used to fetch data

```python
butler = Butler('/repo/main')
```

### The processing runs of the data pipeline are referred to as "collections". You can query what collections are available:

```python
# For a recent processing of the HSC data, choose one of the collections in the format "HSC/runs/RC2/w_YYYY_WW"
collections = butler.registry.queryCollections('*HSC/runs/RC2/w_2023*')
collections
```

```python
butler = Butler('/repo/main', collections=['HSC/runs/RC2/w_2023_01/DM-37483'])
```

### Get the catalogs for sources before global calibration:
(Single-frame measurements are called "sources". The table of all sources from one visit before global calibration is called the "preSourceTable_visit")

```python
# butler.registry.queryDatasets gives you an iterable of the references to the data:
preSource_tables = list(set(butler.registry.queryDatasets("preSourceTable_visit", band='z', tract=9813,
                                                         skymap='hsc_rings_v1')))
```

```python
len(preSource_tables)
```

```python
# butler.get actually retrieves the data:
preSource_table = butler.get(preSource_tables[0])
preSource_table
```

### Get the same sources after global calibration

The data is fetched using the dataset type and the dataId. The dataId is similar to a dictionary with certain dimensions given the dataset type.

```python
# An example dataId
preSource_tables[0].dataId
```

```python
# To get the sources from the table above after calibration, we use the same dataId, but with
# the dataset type name of the calibrated tables:
source_table = butler.get("sourceTable_visit", dataId=preSource_tables[0].dataId)
source_table
```

## Get the objects from coadded images

#### The sky is split into tracts, which are then split into patches. The "skymap" determines how this is done. This is done in order to have manageable amounts of data to process at once. Here we get the objects in one patch of one tract. You can also get the "objectTable_tract", which is for the whole tract, but that is a _very_ large table.

```python
objectTable = butler.get("objectTable", tract=9813, patch=42, band='i', skymap='hsc_rings_v1')
objectTable
```

## Get general information about all the visits

```python
visitTable = butler.get('visitTable')
visitTable
```

## Or get information for all detectors for all visits

```python
ccdVisitTable = butler.get("ccdVisitTable")
ccdVisitTable
```
